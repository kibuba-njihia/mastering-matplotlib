```python
import numpy as np
import matplotlib.pyplot as plt
```

The plot() function by default draws a line from one point to another given inputs. To plot only the markers, you can use 'o'

```python
xpoints = np.array([-3, 6])
ypoints = np.array([50, 250])

plt.plot(xpoints, ypoints,'o')
plt.show()
```

To plot multiple points, you can pass the array of the points. They should be of the same length

```python
x_axis=np.array([1,-3,5,-7])
y_axis=np.array([2,0,0,6])

plt.plot(x_axis,y_axis)
plt.show()
```

If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 etc., depending on the length of the y-points.

```python
x_axis=np.array([2,0,0,6])

plt.plot(x_axis)
plt.show()
```
