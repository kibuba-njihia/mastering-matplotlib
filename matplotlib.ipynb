```python
import numpy as np
import matplotlib.pyplot as plt
```

The plot() function by default draws a line from one point to another given inputs. To plot only the markers, you can use 'o'

```python
xpoints = np.array([-3, 6])
ypoints = np.array([50, 250])

plt.plot(xpoints, ypoints,'o') # .plot is used to create a line plot
plt.show()
```

To plot multiple points, you can pass the array of the points. They should be of the same length

```python
x_axis=np.array([-10,2,5,7])
y_axis=np.array([2,0,0,6])

plt.plot(x_axis,y_axis)
plt.show()
```

If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 etc., depending on the length of the y-points.

```python
x_axis=np.array([2,0,0,6])

plt.plot(x_axis)
plt.show()
```

Markers: To mark specific points on the plot


You can add many more params for colors, line type, etc

```python
x_axis=np.array([2,0,3,6])

plt.plot(x_axis,'o--',marker='h',ms = 10,mec = 'r',c = '#4CAF10',mfc = '#4CAF50')
plt.show()
```

### **Labels and Title**

Use xlabel() and ylabel() functions for labels and title() for the title.

```python
array_one=np.array([1,6])
array_two=np.array([-20,60])

plt.plot(array_one)
plt.plot(array_two)

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.xlabel("x-axis",fontdict=font1,size=12)
plt.ylabel("y-axis",fontdict=font1,size=12)
plt.title("Line Plot",fontdict=font2,size=20,loc='Center')
plt.grid(axis='y',color = 'green', linestyle = '--', linewidth = 0.5) # Adding gridlines
plt.show()
```

### **Display Multiple Plots**

Use subplot() function which takes in 3 args. 1st & 2nd are the rows and columns, 3rd is the index of the plot.

plt.subplot(1, 2, 1) - The figure has 1 row, 2 columns, and this plot is the first plot.

plt.subplot(1, 2, 2) - The figure has 1 row, 2 columns, and this plot is the second plot.

```python
#First plot
a=np.array([10,6])
b=np.array([-20,60])

plt.subplot(1,2,1)
plt.plot(a,b)
plt.title("Loss")

#Second plot
c=np.array([3,8])
d=np.array([1,4])

plt.subplot(1,2,2)
plt.plot(c,d)
plt.title("Income")

plt.suptitle("Trend")
plt.show()
```

### **Scatter plots**

Use .scatter() ftn. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis:

```python
#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color = 'hotpink')

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([10,10,8,10,9,9,9,5,4,10,7,11,1,8,8])
plt.scatter(x, y)
plt.show()
```

####Adding colorbar

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='copper')

plt.colorbar()

plt.show()
```

####Changing size

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes,alpha=0.5)

plt.show()
```

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```
