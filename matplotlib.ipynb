```python
import numpy as np
import matplotlib.pyplot as plt
```

The plot() function by default draws a line from one point to another given inputs. To plot only the markers, you can use 'o'

```python
xpoints = np.array([-3, 6])
ypoints = np.array([50, 250])

plt.plot(xpoints, ypoints,'o') # .plot is used to create a line plot
plt.show()
```

To plot multiple points, you can pass the array of the points. They should be of the same length

```python
x_axis=np.array([1,-3,5,-7])
y_axis=np.array([2,0,0,6])

plt.plot(x_axis,y_axis)
plt.show()
```

If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 etc., depending on the length of the y-points.

```python
x_axis=np.array([2,0,0,6])

plt.plot(x_axis)
plt.show()
```

Markers: To mark specific points on the plot


You can add many more params for colors, line type, etc

```python
x_axis=np.array([2,0,3,6])

plt.plot(x_axis,'o--',marker='h',ms = 10,mec = 'r',c = '#4CAF10',mfc = '#4CAF50')
plt.show()
```

### **Labels and Title**

Use xlabel() and ylabel() functions for labels and title() for the title.

```python
array_one=np.array([1,6])
array_two=np.array([-20,60])

plt.plot(array_one)
plt.plot(array_two)

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.xlabel("x-axis",fontdict=font1,size=12)
plt.ylabel("y-axis",fontdict=font1,size=12)
plt.title("Line Plot",fontdict=font2,size=20,loc='Center')
plt.grid(axis='y',color = 'green', linestyle = '--', linewidth = 0.5) # Adding gridlines
plt.show()
```

### **Display Multiple Plots**

Use subplot() function which takes in 3 args. 1st & 2nd are the rows and columns, 3rd is the index of the plot.

plt.subplot(1, 2, 1) - The figure has 1 row, 2 columns, and this plot is the first plot.

plt.subplot(1, 2, 2) - The figure has 1 row, 2 columns, and this plot is the second plot.

```python
#First plot
a=np.array([10,6])
b=np.array([-20,60])

plt.subplot(1,2,1)
plt.plot(a,b)
plt.title("Loss")

#Second plot
c=np.array([3,8])
d=np.array([1,4])

plt.subplot(1,2,2)
plt.plot(c,d)
plt.title("Income")

plt.suptitle("Trend")
plt.show()
```
